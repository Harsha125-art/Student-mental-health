```python
import pandas as pd
df = pd.read_csv('/kaggle/input/student-mental-health/Student Mental health.csv')
```

```python
df.head()
```

Renaming The columns 

```python
df.columns = df.columns.str.strip().str.replace(' ','_').str.lower()
df= df.rename(columns={'choose_your_gender':'gender','what_is_your_course?':'course','your_current_year_of_study':'study_year','what_is_your_cgpa?':'cgpa','do_you_have_depression?':'depression','do_you_have_anxiety?':'anxiety','do_you_have_panic_attack?':'panic_attack','did_you_seek_any_specialist_for_a_treatment?':'treatment'})
df.head()
```

There are Y and small y is year in some rows convert them to 'y' for clear visualization

```python
df['study_year'] = df['study_year'].str.strip().str.lower()
df.head()
```

```python
df.to_csv('cleaned_student_mental_health_dataset.',index=False)
```
